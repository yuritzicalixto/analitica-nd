```python
import pandas as pd
```

```python
pokemon = pd.read_csv('pokemon.csv')
```

```python
pokemon
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Graficando las estadisticas
# Hisplot es una funciÃ³n de Seaborn que crea un histograma: un grafico de barras
# que muestra la frecuencia con la que aparecen los valores de una variable
# sns.histplot(data=df, x='NombreDeLaColumna')
sns.histplot(pokemon, x="Generation")
plt.title("Estadisticas de Pokemon")
plt.xlabel("GeneraciÃ³n")
plt.ylabel("Count")
plt.show()
```

```python
# bins=grosor de la linea, True = la curva
sns.histplot(pokemon, x='Attack', bins=20, kde=True)
plt.title('DistribuciÃ³n del Ataque con 20 bins')
plt.show()
```

```python
sns.countplot(pokemon["Type 1"])
```

```python
sns.countplot(pokemon, x="Type 1", hue="Generation")
```

```python
# Colorea si es necesario
sns.histplot(data=pokemon, x="Type 2", hue="Legendary")
plt.title("Total de EstadÃ­sticas por Tipo de PokÃ©mon")
plt.xticks(rotation=45)#Rota las etiquetas del eje X
plt.xlabel("Tipo de pokemon")
plt.ylabel("NÃºmero de PokÃ©mon")
plt.show()
```

```python
# Datos atipicos
# ScatterPlot
# Un Scatterplot (grÃ¡fico de dispersiÃ³n) muestra puntos individuales en el plano cartesiano
sns.scatterplot(data=pokemon, x='Attack', y='Defense')
plt.title('Ataque vs Defensa')
plt.xlabel('Ataque')
plt.ylabel('Defensa')
plt.show()
```

```python
# Colorear segÃºn si es legendario
sns.scatterplot(pokemon, x='Attack', y='Defense', hue='Legendary')
plt.title('Ataque vs Defensa (Legendarios vs Normales)')
plt.show()
```

```python
# Cambiar tamaÃ±o de los puntos segÃºn la velocidad
sns.scatterplot(pokemon, x='Attack', y='Defense', size='Speed',
        hue='Legendary', alpha=0.8)
plt.title('Ataque vs Defensa (TamaÃ±o segÃºn velocidad)')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```
