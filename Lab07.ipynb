```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Distribuci칩n de calificaciones IMDb
df = pd.read_csv('Movie.csv')
df
# Grafica la distribuci칩n de la variable IMDB_Rating con un histograma
# Incluye el par치metro kde=True para ver la densidad
plt.figure(figsize=(8, 5))
sns.histplot(df['Rating'], kde=True, bins=20)
plt.title('Distribuci칩n de calificaciones IMDb')
plt.xlabel('IMDb Rating')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Crea un scatterplot entre IMDB_Rating y Revenue_Million
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Rating', y='Revenue (Millions)')
plt.title('IMDb Rating vs Revenue (Millions)')
plt.xlabel('IMDb Rating')
plt.ylabel('Revenue (Millions)')
plt.show()
```

```python
# Crea un boxplot para ver la distribuci칩n de Runtime por cada Genre
df['Main_Genre'] = df['Genre'].apply(lambda x: x.split(',')[0] if pd.notnull(x) else x)
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Main_Genre', y='Runtime (Minutes)')
plt.title('Distribuci칩n de Runtime por G칠nero Principal')
plt.xlabel('G칠nero')
plt.ylabel('Duraci칩n (minutos)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Realiza un violinplot que compare IMDB_Rating por g칠nero
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='Main_Genre', y='Rating')
plt.title('Distribuci칩n de IMDb Rating por G칠nero Principal')
plt.xlabel('G칠nero')
plt.ylabel('IMDb Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
#Crea un scatterplot de IMDB_Rating vs Revenue_Million y usa hue='Year'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Rating', y='Revenue (Millions)', hue='Year', palette='viridis')
plt.title('IMDb Rating vs Revenue_Millions por A침o')
plt.xlabel('IMDb Rating')
plt.ylabel('Revenue (Millions)')
plt.legend(title='A침o', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
