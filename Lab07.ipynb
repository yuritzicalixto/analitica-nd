```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Distribución de calificaciones IMDb
df = pd.read_csv('Movie.csv')
df
# Grafica la distribución de la variable IMDB_Rating con un histograma
# Incluye el parámetro kde=True para ver la densidad
plt.figure(figsize=(8, 5))
sns.histplot(df['Rating'], kde=True, bins=20)
plt.title('Distribución de calificaciones IMDb')
plt.xlabel('IMDb Rating')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Crea un scatterplot entre IMDB_Rating y Revenue_Million
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Rating', y='Revenue (Millions)')
plt.title('IMDb Rating vs Revenue (Millions)')
plt.xlabel('IMDb Rating')
plt.ylabel('Revenue (Millions)')
plt.show()
```

```python
# Crea un boxplot para ver la distribución de Runtime por cada Genre
df['Main_Genre'] = df['Genre'].apply(lambda x: x.split(',')[0] if pd.notnull(x) else x)
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Main_Genre', y='Runtime (Minutes)')
plt.title('Distribución de Runtime por Género Principal')
plt.xlabel('Género')
plt.ylabel('Duración (minutos)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Realiza un violinplot que compare IMDB_Rating por género
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='Main_Genre', y='Rating')
plt.title('Distribución de IMDb Rating por Género Principal')
plt.xlabel('Género')
plt.ylabel('IMDb Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
#Crea un scatterplot de IMDB_Rating vs Revenue_Million y usa hue='Year'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Rating', y='Revenue (Millions)', hue='Year', palette='viridis')
plt.title('IMDb Rating vs Revenue_Millions por Año')
plt.xlabel('IMDb Rating')
plt.ylabel('Revenue (Millions)')
plt.legend(title='Año', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
