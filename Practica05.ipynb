```python
# Importar las librerias necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfNetflix = pd.read_csv('netflix_titles.csv')
```

```python
# Estadisticas basicas
dfNetflix.describe()
```

```python
# Limpieza de datos ¿como se buscan valores nulos?
dfNetflix.isnull().sum()
```

```python
#Eliminar datos nulos, aqui se crea primero una copia
dfLimpio = dfNetflix.dropna(subset=['country', 'cast', 'date_added'])
```

```python
#Validar
dfLimpio.isnull().sum()
```

```python
# Rellenar la información
dfLimpio['director']=dfLimpio['director'].fillna('Desconocido')
```

```python
dfLimpio.isnull().sum()
```

```python
# Analisis Descriptivo
# Mostrar cuentos registros existen por tipo, value_counts dime cuantos hay
dfLimpio['type'].value_counts()
```

```python
# Graficaremos la info anterior Set2=rojo y azul, legend=False
sns.countplot(data=dfLimpio, x='type', hue='type', palette='Set2', legend=True)
#A la grafica resultante le pondremos un titulo
plt.title('Total por tipo de contenido')
plt.show()
```

```python
#
sns.countplot(data=dfLimpio, x='country', hue='country', palette='Set2')
plt.title('Clasificación por pais')
plt.show()
```

```python
# Grafica de los primeros 10
topPaises = dfLimpio['country'].value_counts().head(10)
topPaises.plot(kind='bar',color='skyblue')
plt.title('10 Países con mayor contenido')
plt.ylabel('Número de titulos ')
plt.show()
```

```python
# Mostrar el contenido por el año de lanzamiento
dfLimpio['release_year'].value_counts().sort_index().plot(kind='line')
plt.title('Contenido agregado pro año')
plt.xlabel('Año')
plt.grid(True)
plt.show()
```

```python
# Mostrar las peliculas
dfMovies=dfLimpio[dfLimpio['type']=='Movie']
```

```python
dfMovies
```

```python
#Graficar peliculas por año
dfMovies['release_year'].value_counts().sort_index().plot(kind='pie')
plt.title('Contenido agregado por año')
plt.xlabel('Año')
plt.grid(True)
plt.show()
```

```python
# Correlaciones
sns.heatmap(dfMovies.corr(numeric_only=True), annot=True)
plt.title('Mapa de calor de correlaciones')
plt.show()
```

```python
# Mostrar los 10 directores con más peliculas
dfTopDirector = dfMovies['director'].value_counts().head(10)
```

```python
dfTopDirector
```

```python
sns.barplot(x=dfTopDirector.values, y=dfTopDirector.index, hue=dfTopDirector.index,palette='viridis')
plt.xlabel('Número de peliculas')
plt.ylabel('Director')
plt.title('Top Directores')
plt.show()
```
